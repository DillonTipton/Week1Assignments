```python
from pyspark.sql import SparkSession, functions as func
from pyspark.sql.types import StructType, StructField, StringType


spark = SparkSession.builder.appName("IPinfo").getOrCreate()

# access_log.txt
inputDF = spark.read.text("access_log.txt")

# Splitting by spaces
df2 = inputDF.withColumn("parts", func.split("value", " "))

# column 0 = IP address
# column 6 = endpoint
# column 8 = status
df3 = (
    df2.withColumn("IP_address", func.col("parts")[0])
       .withColumn("field1", func.col("parts")[1])
       .withColumn("field2", func.col("parts")[2])
        .withColumn("field3", func.col("parts")[3])
       .withColumn("field4", func.col("parts")[4])
       .withColumn("field5", func.col("parts")[5])
       .withColumn("endpoint", func.col("parts")[6])
        .withColumn("field7", func.col("parts")[7])
       .withColumn("status", func.col("parts")[8])
       .withColumn("field9", func.col("parts")[9])
)
df3 = df3.drop("value")
df3 = df3.drop("parts")
df3.show()

ip_count = df3.groupBy("IP_address").agg(func.count('*').alias('request_count')).sort("request_count", ascending=False)
ip_count.show()

endpoint_count = df3.groupBy("endpoint").agg(func.count('*').alias('endpoint_count')).sort("endpoint_count", ascending=False)
endpoint_count.show()

status_count = df3.groupBy("status").agg(func.count('*').alias('status_count')).sort("status_count", ascending=False)
status_count.show()

spark.stop()
```
